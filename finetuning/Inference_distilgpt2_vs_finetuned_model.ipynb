```python
from transformers import AutoTokenizer, AutoModelForCausalLM
tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
model = AutoModelForCausalLM.from_pretrained("distilgpt2")
```

```python
promt_text = """
Ingredients:
yellow cake mix
water
vegetable oil
eggs
creamy peanut butter
chocolate frosting
peanuts
Instructions:
"""
```

```python
encoded_promt = tokenizer.encode(
    promt_text, 
    add_special_tokens=False, 
    return_tensors='pt'
)

encoded_promt
```

```python
output_sequences = model.generate(
    input_ids=encoded_promt,
    max_length=600,
    temperature=0.9,
    top_k=20,
    top_p=0.9,
    repetition_penalty=1,
    do_sample=True,
    num_return_sequences=4,
    early_stopping=True
)
```

```python
special_token = '<|endoftext|>'
```

```python
for i, output_sequence in enumerate(output_sequences):
  result = tokenizer.decode(output_sequence)
  print('-------', i)
  print(result)
```

##Load finetuned model

```python
custom_tokenizer = AutoTokenizer.from_pretrained('/content/drive/MyDrive/ml/recipes/backup_data/epochs_2')
```

```python
custom_model = AutoModelForCausalLM.from_pretrained('/content/drive/MyDrive/ml/recipes/backup_data/epochs_2')
```

```python
encoded_promt_finetuned = custom_tokenizer.encode(
    promt_text, 
    add_special_tokens=False, 
    return_tensors='pt'
)
```

```python
output_sequences_finetuned = custom_model.generate(
    input_ids=encoded_promt_finetuned,
    max_length=600,
    temperature=0.9,
    top_k=20,
    top_p=0.9,
    repetition_penalty=1,
    do_sample=True,
    num_return_sequences=4,
    early_stopping=True
)
```

```python
for i, output_sequence in enumerate(output_sequences_finetuned):
  result = custom_tokenizer.decode(output_sequence)
  result = result[:result.index(special_token)]
  print(result)
  print('----')
```
